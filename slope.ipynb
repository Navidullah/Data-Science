```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Example data: x and y values
x = np.array([30,31, 32, 34, 36, 38,39,40,42])
y = np.array([220,230, 240,250,260,270,280,290,300])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Generate regression line values
regression_line = slope * x + intercept

# Plot the data points
plt.scatter(x, y, color='blue', label='Data Points')

# Plot the regression line
plt.plot(x, regression_line, color='red', label=f'Line: y={slope:.2f}x+{intercept:.2f}')

# Annotate the slope and intercept
##plt.text(0.5, 4.8, f'Slope: {slope:.2f}', fontsize=12, color='red')
#plt.text(0.5, 4.4, f'Y-intercept: {intercept:.2f}', fontsize=12, color='red')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0,44)
plt.ylim(0,330)
plt.title('Linear Regression')
plt.legend()

# Show plot
plt.show()
```
